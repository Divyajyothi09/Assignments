```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('my_movies.csv')
df
```

```python
df.info()
```

```python
df.shape
```

```python
movies = []
```

```python
for i in range(0, 10):  
  movies.append([str(df.values[i,j]) for j in range(0, 15)]) 
movies
```

```python
from apyori import apriori
rules = apriori(movies, min_support = 0.20,min_confidence = 0.02,min_lift = 1,min_length=2)
rules=list(rules)
print(rules)
```

```python
print(len(rules))
```

```python
support=[]
confidence = []
lift = []
```

```python
for i in range(0,87):       
    support.append(rules[i][1])
    confidence.append(rules[i][2][0][2])
    lift.append(rules[i][2][0][3])
```

```python
d1=pd.DataFrame(support,columns=['Support'])
d2=pd.DataFrame(confidence,columns=['Confidence'])
d3=pd.DataFrame(lift,columns=['Lift'])
```

```python
rules = pd.concat([d1,d2,d3],axis=1)
rules
```

```python
plt.scatter(rules['Support'],rules['Confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

```python
sns.heatmap(data=rules)
plt.show()
```
